## Explain dataset

Reference: https://rstudio-pubs-static.s3.amazonaws.com/155304_cc51f448116744069664b35e7762999f.html

* id: unique id of home sold
* date: date of the home sale
* price: price of each home sold
* bedrooms: number of bedrooms
* bathrooms: number of bathrooms: .25: toilet only, .5: bathroom with toilet and no shower, .75: sink, toilet, and without tub or shower
* sqft_living: square footage of the apartments interior living space
* sqft_lot: square footage of the land space
* floors: number of floors
* waterfront: whether the apartment was overlooking the river or not (0: no, 1: yes)
* view: 0 to 4 to indicate how good the view was
* condition: 1 to 5 to show the condition of the apartment
* grade: 1 to 13. 1-3: short of building construction and design, 7: average level of construction and design, 11-13: high quality level of construction and design.
* sqft_above: square footage of the interior housing space that is above ground level
* sqft_basement: square footage of the interior housing space that is below ground level
* yr_built: year the house was initially built
* yr_renovated: year of the houseâ€™s last renovation
* zipcode: zipcode area the house is in
* lat: lattitude
* long: longitude
* sqft_living15: square footage of interior housing living space for the nearest 15 neighbors
* sqft_lot15: square footage of the land lots of the nearest 15 neighbors

```python
newdf['construction_design_level']=newdf['grade']
```

```python
newdf['construction_design_level'].value_counts()
```

```python
#1->3: low
#4->6: fair
#7: average
#8-10: good
#11-13: high quality
def categorizeGrade(grade):
    if grade>=1 & grade<=3: return "Low"
    elif grade>=4 & grade<=6: return "Fair"
    elif grade==7: return "Average"
    elif grade>=8 & grade<=10: return "Good"
    elif grade>=11 & grade<=13: return "High"
```

```python
newdf['construction_design_level']=newdf['grade'].apply(lambda x: categorizeGrade(x))
```
